```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
products_df = pd.read_csv("../data/amazon_products.csv")
```

```python
products_df.head(5)
```

```python
unwanted_columns = ['productURL']
products_df = products_df.drop(unwanted_columns, axis=1)
```

```python
products_df.head(5)
```

```python
products_df.describe()
```

```python
category_df = pd.read_csv("../data/amazon_categories.csv")
category_df.head(10)
```

```python
final_df = pd.merge(products_df, category_df, how='inner',left_on='category_id', right_on='id')
```

```python
final_df.head(5)
```

```python
final_df['title'] = final_df['title'].fillna('Unknown Title')
```

```python
unwanted_columns = ["asin", "title","imgUrl", "id", "category_name"]
all_num_df = final_df.drop(unwanted_columns, axis=1)
correlation_matrix = all_num_df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Feature Correlation heatmap")
plt.show()
```

```python
#Find the most selling product of each category
final_df['short_title'] = final_df['title'].apply(lambda x: x[:30] + "...." if len(x) > 30 else x)
most_sold_products = final_df.loc[final_df.groupby('category_id')['boughtInLastMonth'].idxmax()]

print(most_sold_products[['short_title', 'boughtInLastMonth']].head(5))
```

```python
most_sold_products.head(10)[['short_title', 'boughtInLastMonth']].set_index('short_title').plot(kind='barh', legend=False)
plt.xlabel('Units Sold in Last Month')
plt.ylabel('Product title')
plt.title('Most Sold Products in Last Month')
plt.show()
```

```python
#Find the top 10 best rated products of each category

top_rated_products = final_df.groupby('category_id').apply(lambda group : group.sort_values('stars', ascending=False).head(10))
print(top_rated_products[["short_title","category_name"]].head(10))
```
