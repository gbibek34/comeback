```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

```python
products_df = pd.read_csv("../data/amazon_products.csv")
```

```python
products_df.head(5)
```

```python
products_df.info()
```

```python
unwanted_columns = ['productURL']
products_df = products_df.drop(unwanted_columns, axis=1)
```

```python
products_df.head(10)
```

```python
category_df = pd.read_csv("../data/amazon_categories.csv")
category_df.head(10)
```

```python
final_df = pd.merge(products_df, category_df, how='inner', left_on='category_id', right_on='id')
```

```python
final_df.head(5)
```

```python
unwanted_columns = ['asin','title','imgUrl', 'id', 'category_name']
all_num_df = final_df.drop(unwanted_columns, axis=1)
correlation_matrix = all_num_df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, square=True)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
